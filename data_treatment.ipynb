```python
DATA_DIR = "condition+monitoring+of+hydraulic+systems"

# Fichiers à exclure pour le chargement 
EXCLUDED_FILES = {
    "profile.txt",
    "description.txt",
    "documentation.txt"
}

# Fonction de chargement des fichiers de capteurs
def load_and_rename_sensor(file_path):
    """
    Charge un fichier capteur (PS1, FS1, TS1, etc.)
    et renomme les colonnes : NomCapteur_1, NomCapteur_2, ...
    """
    sensor_name = os.path.splitext(os.path.basename(file_path))[0]

    df = pd.read_csv(
        file_path,
        sep=r"\s+",
        header=None,
        encoding="latin1"
    )

    df.columns = [f"{sensor_name}_{i+1}" for i in range(df.shape[1])]
    return df

# Chargement des fichiers capteurs uniquement
sensor_dfs = []

for file in sorted(os.listdir(DATA_DIR)):
    file_lower = file.lower()

    if (
        file_lower.endswith(".txt")
        and file_lower not in EXCLUDED_FILES
    ):
        file_path = os.path.join(DATA_DIR, file)
        df_sensor = load_and_rename_sensor(file_path)
        sensor_dfs.append(df_sensor)
        print(f"✔ Chargé : {file}")

# Chargement du fichier profile contenant la variable cible
profile_df = pd.read_csv(
    os.path.join(DATA_DIR, "profile.txt"),
    sep=r"\s+",
    header=None,
    encoding="latin1"
)

profile_df.columns = [
    "cooler_condition",
    "valve_condition",
    "internal_pump_leakage",
    "hydraulic_accumulator",
    "stable_flag"
]

# Fusion finale
final_df = pd.concat(sensor_dfs + [profile_df], axis=1)

print("\n Fusion terminée")
print("Shape finale :", final_df.shape)
print("\nDernières colonnes (targets) :")
print(final_df.columns[-5:])
```

```python
# Détection des outliers (méthode IQR – globale)
numeric_df = final_df.select_dtypes(include=[np.number])

outliers_summary = []

for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1

    if IQR == 0:
        continue

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers_count = ((numeric_df[col] < lower) | (numeric_df[col] > upper)).sum()

    if outliers_count > 0:
        outliers_summary.append({
            "variable": col,
            "outliers_count": outliers_count,
           # "outliers_percent": (outliers_count / len(numeric_df)) * 100
        })

outliers_df = pd.DataFrame(outliers_summary)

print("\nNombre de variables contenant des outliers :", outliers_df.shape[0])
print(outliers_df.head())
```

```python
# Identification des outliers PAR CAPTEUR
print("RÉSUMÉ DES OUTLIERS PAR CAPTEUR")

# Extraire le nom du capteur (avant le "_")
outliers_df["capteur"] = outliers_df["variable"].str.extract(r"^([A-Z0-9]+)_", expand=False)

# Grouper par capteur et calculer le nombre total d'outliers
capteur_summary = outliers_df.groupby("capteur").agg({
    "variable": "count",
    "outliers_count": "sum"
}).round(2)

capteur_summary.columns = ["nb_variables_avec_outliers", "total_outliers"]
capteur_summary = capteur_summary.sort_values("total_outliers", ascending=False)

print("\n", capteur_summary)

# Visualisation
plt.figure(figsize=(12, 6))
capteur_summary["total_outliers"].plot(kind="bar", color="steelblue")
plt.title("Nombre total d'outliers par capteur")
plt.ylabel("Nombre d'outliers")
plt.xlabel("Capteur")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.show()

print("\n Les outliers sont présents dans : ", list(capteur_summary.index))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print("CORRÉLATION : OUTLIERS vs VARIABLES CIBLES")

# Colonnes cibles
target_cols = [
    "cooler_condition",
    "valve_condition",
    "internal_pump_leakage",
    "hydraulic_accumulator",
    "stable_flag"
]

# Création des indicateurs d’outliers
outliers_flags = numeric_df.copy()

for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1

    if IQR == 0:
        outliers_flags[col] = 0
        continue

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers_flags[col] = (
        (numeric_df[col] < lower) | (numeric_df[col] > upper)
    ).astype(int)

# Ajouter les cibles
for col in target_cols:
    if col in final_df.columns:
        outliers_flags[col] = final_df[col]

# Calcul des corrélations
correlations = []

for var in outliers_df["variable"]:
    for target in target_cols:
        if target in outliers_flags.columns:
            corr = outliers_flags[var].corr(outliers_flags[target])
            correlations.append({
                "variable_outliers": var,
                "cible": target,
                "correlation": corr,
                "abs_correlation": abs(corr)
            })

corr_df = pd.DataFrame(correlations)

strong_corr = (
    corr_df[corr_df["abs_correlation"] > 0.1]
    .sort_values("abs_correlation", ascending=False)
)

print("\nCorrélations FORTES (|r| > 0.1) :")
print(strong_corr.head(15).to_string())


if strong_corr.empty:
    print("\nAucune corrélation forte détectée")
    print("→ Les outliers n’impactent pas directement les cibles")
else:
    print("\nCorrélations significatives détectées")
    print("→ Les outliers peuvent contenir de l'information utile")


plt.figure(figsize=(14, 5))

# --- Graphique 1 : distribution des corrélations ---
plt.subplot(1, 2, 1)
plt.hist(corr_df["abs_correlation"], bins=30)
plt.axvline(0.1, linestyle="--", linewidth=2, label="Seuil (0.1)")
plt.xlabel("Corrélation absolue")
plt.ylabel("Fréquence")
plt.title("Distribution des corrélations Outliers ↔ Cibles")
plt.legend()

# --- Graphique 2 : impact par cible ---
plt.subplot(1, 2, 2)
if not strong_corr.empty:
    top_by_target = (
        strong_corr
        .groupby("cible")["abs_correlation"]
        .max()
        .sort_values()
    )
    plt.barh(top_by_target.index, top_by_target.values)
    plt.xlabel("Corrélation maximale")
    plt.title("Impact maximal des outliers par cible")
else:
    plt.text(0.5, 0.5, "Aucune corrélation forte détectée",
             ha="center", va="center", transform=plt.gca().transAxes)
    plt.axis("off")

plt.show()
```

```python
# COMPARAISON VALEURS NORMALES vs OUTLIERS
print("ANALYSE DÉTAILLÉE : VALEURS NORMALES vs OUTLIERS")

# Sélectionner les 3 capteurs les plus impactés
top_sensors = (
    outliers_df.sort_values("outliers_count", ascending=False)
    ["variable"]
    .head(3)
)

print(f"\nCapteurs analysés : {list(top_sensors)}\n")

def compute_iqr_bounds(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return Q1 - 1.5 * IQR, Q3 + 1.5 * IQR


def split_normal_outliers(series, lower, upper):
    normal = series[(series >= lower) & (series <= upper)]
    outliers = series[(series < lower) | (series > upper)]
    return normal, outliers

results = []

for col in top_sensors:
    lower, upper = compute_iqr_bounds(numeric_df[col])
    normal_vals, outlier_vals = split_normal_outliers(numeric_df[col], lower, upper)

    results.append({
        "Capteur": col,
        "Normal_count": len(normal_vals),
        "Normal_min": normal_vals.min(),
        "Normal_max": normal_vals.max(),
        "Normal_mean": normal_vals.mean(),
        "Normal_std": normal_vals.std(),
        "Outliers_count": len(outlier_vals),
        "Outliers_min": outlier_vals.min(),
        "Outliers_max": outlier_vals.max(),
        "Outliers_mean": outlier_vals.mean(),
        "Outliers_std": outlier_vals.std(),
        "Écart_min (%)": (
            (outlier_vals.min() - normal_vals.min()) / normal_vals.min() * 100
            if normal_vals.min() != 0 else np.nan
        ),
        "Écart_max (%)": (
            (outlier_vals.max() - normal_vals.max()) / normal_vals.max() * 100
            if normal_vals.max() != 0 else np.nan
        )
    })

comparison_df = pd.DataFrame(results)
print(comparison_df.T.to_string())

for col in top_sensors:
    lower, upper = compute_iqr_bounds(numeric_df[col])
    normal_vals, outlier_vals = split_normal_outliers(numeric_df[col], lower, upper)

    plt.figure(figsize=(16, 4))

    plt.subplot(1, 3, 1)
    plt.hist(normal_vals, bins=30, alpha=0.7, label="Normal")
    plt.hist(outlier_vals, bins=30, alpha=0.7, label="Outliers")
    plt.title(f"{col} - Distribution")
    plt.xlabel("Valeur")
    plt.ylabel("Fréquence")
    plt.legend()
    plt.grid(alpha=0.3)

    plt.subplot(1, 3, 2)
    plt.boxplot([normal_vals, outlier_vals], labels=["Normal", "Outliers"])
    plt.title(f"{col} - Boxplot")
    plt.ylabel("Valeur")
    plt.grid(axis="y", alpha=0.3)

    plt.subplot(1, 3, 3)
    sample_idx = np.random.choice(len(numeric_df), min(200, len(numeric_df)), replace=False)
    sample_data = numeric_df[col].iloc[sample_idx]
    is_outlier = (sample_data < lower) | (sample_data > upper)

    plt.scatter(
        range(len(sample_data)),
        sample_data,
        c=is_outlier.map({True: "red", False: "green"}),
        alpha=0.6,
        s=20
    )
    plt.axhline(lower, linestyle="--", label="Seuil bas")
    plt.axhline(upper, linestyle="--", label="Seuil haut")
    plt.title(f"{col} - Échantillon temporel")
    plt.xlabel("Échantillon")
    plt.ylabel("Valeur")
    plt.legend(fontsize=8)
    plt.grid(alpha=0.3)

    plt.show()

print("\nAnalyse terminée")
```

```python
# LIEN ENTRE OUTLIERS ET CONDITIONS DES COMPOSANTS
print("ANALYSE : OUTLIERS selon l'état des COMPOSANTS")

# Sélectionner un capteur représentatif
sensor_col = top_sensors.iloc[0]

# Calcul des bornes IQR
Q1 = numeric_df[sensor_col].quantile(0.25)
Q3 = numeric_df[sensor_col].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Créer un indicateur outlier
final_df['is_outlier'] = ((numeric_df[sensor_col] < lower) | (numeric_df[sensor_col] > upper)).astype(int)

# Analyser la distribution des outliers par condition
print(f"\nAnalyse sur le capteur : {sensor_col}\n")

target_cols = ["cooler_condition", "valve_condition", "internal_pump_leakage", "hydraulic_accumulator"]

for target in target_cols:
    print(f"{target.upper()}")
    
    # Tableau croisé en pourcentage
    cross_tab = pd.crosstab(final_df[target], final_df['is_outlier'], normalize='index') * 100
    cross_tab.columns = ['Normal (%)', 'Outlier (%)']
    print("\n", cross_tab.round(2))
    
    # Compter le nombre d'outliers et de normaux par condition
    outlier_count = final_df[final_df['is_outlier'] == 1].groupby(target).size()
    normal_count = final_df[final_df['is_outlier'] == 0].groupby(target).size()
    
    summary = pd.DataFrame({
        'Normal': normal_count,
        'Outliers': outlier_count,
        'Total': normal_count + outlier_count,
        '% Outliers': (outlier_count / (normal_count + outlier_count) * 100).round(2)
    }).fillna(0)
    
    print("\nRépartition détaillée :")
    print(summary)
    
    # Visualisation simple avec plt
    plt.figure(figsize=(8,5))
    summary[['Normal','Outliers']].plot(kind='bar', stacked=False, color=['green', 'red'], alpha=0.7)
    plt.title(f"Outliers selon {target}")
    plt.xlabel(target)
    plt.ylabel("Nombre d'observations")
    plt.xticks(rotation=0)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

# Supprimer la colonne temporaire
final_df.drop(columns=['is_outlier'], inplace=True)

print("\nLes outliers indiquent potentiellement des défauts/dégradations !")
```

```python
# Préparation des features et de la cible
print("PRÉPARATION DES FEATURES ET DE LA CIBLE")

# Séparer les features des cibles
target_columns = [
    "cooler_condition", 
    "valve_condition", 
    "internal_pump_leakage", 
    "hydraulic_accumulator", 
    "stable_flag",
    "valve_binary"
]

# Features = toutes les colonnes sauf les cibles
X = final_df.drop(columns=target_columns)
y = final_df['valve_binary']

print(f"\n Features (X) : {X.shape}")
print(f"   - {X.shape[0]} observations")
print(f"   - {X.shape[1]} features (capteurs)")

print(f"\n Cible (y) : {y.shape}")
print(f"   - Classes : {y.unique()}")
print(f"   - Distribution : {dict(y.value_counts())}")
```

```python
# NORMALISATION avec RobustScaler (résistant aux outliers)
from sklearn.preprocessing import RobustScaler

print("NORMALISATION AVEC ROBUSTSCALER")
print("\nRobustScaler utilise la médiane et l'IQR")
print("Résistant aux outliers")
print("Préserve l'information des valeurs extrêmes\n")

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)

# Convertir en DataFrame pour garder les noms de colonnes
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)

print(f"Normalisation terminée")
print(f"Shape : {X_scaled_df.shape}")

# Visualisation avant/après normalisation (échantillon)
sample_features = X.columns[:3]  # 3 premières features

for feature in sample_features:
    plt.figure(figsize=(7,4))
    
    # Avant normalisation
    plt.hist(X[feature], bins=50, color='steelblue', alpha=0.7, label='Avant')
    
    # Après normalisation
    plt.hist(X_scaled_df[feature], bins=50, color='green', alpha=0.5, label='Après')
    
    plt.title(f"Comparaison AVANT/APRÈS - {feature}")
    plt.xlabel("Valeur")
    plt.ylabel("Fréquence")
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

print("\nFeatures normalisées : X_scaled_df")
```

```python
print("SPLIT TRAIN/TEST")

print("TRAIN : Cycles SANS outliers (données de référence saines)")
print("TEST  : Cycles AVEC outliers (anomalies à détecter)")
print("Permet au modèle d'apprendre le comportement normal")
print("Le test évalue la capacité à détecter les défauts\n")


print("Étape 1 : Détection des cycles contenant des outliers...\n")

has_outlier_mask = pd.Series(False, index=X.index)

for col in X.columns:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    
    if IQR == 0:
        continue
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    has_outlier_mask |= (X[col] < lower) | (X[col] > upper)

cycles_normal = (~has_outlier_mask).sum()
cycles_anomaly = has_outlier_mask.sum()

print(f"Cycles NORMAUX (sans outliers) : {cycles_normal}")
print(f"Cycles ANOMALIES (avec outliers) : {cycles_anomaly}")
print(f"Total : {len(X)}\n")

print("Étape 2 : Séparation Train/Test...\n")

normal_indices = X.index[~has_outlier_mask]
anomaly_indices = X.index[has_outlier_mask]

X_train = X_scaled_df.loc[normal_indices].copy()
y_train = y.loc[normal_indices].copy()

X_test = X_scaled_df.loc[anomaly_indices].copy()
y_test = y.loc[anomaly_indices].copy()

print(f"ENSEMBLE D'ENTRAÎNEMENT (Train - Cycles Normaux)")
print(f" - X_train : {X_train.shape}")
print(f" - y_train : {y_train.shape}")
print(f" - Distribution : {dict(y_train.value_counts())}")
print(f" - Proportions : {dict(y_train.value_counts(normalize=True).round(3))}")

print(f"\nENSEMBLE DE TEST (Test - Cycles Anomalies)")
print(f" - X_test : {X_test.shape}")
print(f" - y_test : {y_test.shape}")
print(f" - Distribution : {dict(y_test.value_counts())}")
print(f" - Proportions : {dict(y_test.value_counts(normalize=True).round(3))}")


# Distribution globale
plt.figure(figsize=(6,4))
y.value_counts().plot(kind='bar', color=['red', 'green'], alpha=0.7, edgecolor='black')
plt.title(f"Distribution TOTALE (n={len(y)})", fontweight='bold')
plt.xlabel("Classe")
plt.ylabel("Nombre d'observations")
plt.xticks([0,1], ['Non-optimal', 'Optimal'], rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.show()

# Distribution Train
plt.figure(figsize=(6,4))
y_train.value_counts().plot(kind='bar', color=['red', 'green'], alpha=0.7, edgecolor='black')
plt.title(f"TRAIN (Cycles Normaux - n={len(y_train)})", fontweight='bold', color='green')
plt.xlabel("Classe")
plt.ylabel("Nombre d'observations")
plt.xticks([0,1], ['Non-optimal', 'Optimal'], rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.show()

# Distribution Test
plt.figure(figsize=(6,4))
y_test.value_counts().plot(kind='bar', color=['red', 'green'], alpha=0.7, edgecolor='black')
plt.title(f"TEST (Cycles Anomalies - n={len(y_test)})", fontweight='bold', color='red')
plt.xlabel("Classe")
plt.ylabel("Nombre d'observations")
plt.xticks([0,1], ['Non-optimal', 'Optimal'], rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.show()

# Split Train/Test (nombre de cycles)
plt.figure(figsize=(5,4))
plt.bar(['Cycles Normaux', 'Cycles Anomalies'], [cycles_normal, cycles_anomaly], 
        color=['green', 'red'], alpha=0.7, edgecolor='black', linewidth=2)
plt.title("Split Train/Test : Nombre de cycles", fontweight='bold')
plt.ylabel("Nombre de cycles")
plt.grid(axis='y', alpha=0.3)
plt.show()

# Composition des classes - Train
plt.figure(figsize=(5,5))
y_train.value_counts(normalize=True).plot(
    kind='pie', autopct='%1.1f%%', colors=['red','green'], startangle=90,
    title='TRAIN : Composition des classes'
)
plt.ylabel("")
plt.show()

# Composition des classes - Test
plt.figure(figsize=(5,5))
y_test.value_counts(normalize=True).plot(
    kind='pie', autopct='%1.1f%%', colors=['red','green'], startangle=90,
    title='TEST : Composition des classes'
)
plt.ylabel("")
plt.show()

print("\nSplit réalisé avec succès selon l'énoncé du projet !")
print("Les modèles apprendront le comportement NORMAL")
print("Et seront testés sur les ANOMALIES")
print("\nDonnées prêtes pour l'entraînement des modèles !")
```

```python
# RANDOM FOREST + BIAIS
import numpy as np
from sklearn.model_selection import learning_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Entraîner Random Forest
rf_model = RandomForestClassifier(
    n_estimators=100, max_depth=15, random_state=42, n_jobs=-1
)
rf_model.fit(X_train, y_train)

# Accuracy
train_acc = accuracy_score(y_train, rf_model.predict(X_train))
test_acc = accuracy_score(y_test, rf_model.predict(X_test))

print("RANDOM FOREST")
print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# Courbe d'apprentissage - Biais/Variance
train_sizes, train_scores, val_scores = learning_curve(
    rf_model, X_train, y_train, cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy', n_jobs=-1
)

train_mean = np.mean(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Train', linewidth=2)
plt.plot(train_sizes, val_mean, 'o-', color='red', label='Validation', linewidth=2)
plt.title("Random Forest - Courbe d'Apprentissage")
plt.xlabel("Nombre d'exemples d'entraînement")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Biais et Variance
biais = 1 - val_mean
variance = train_mean - val_mean

plt.figure(figsize=(8,5))
plt.plot(train_sizes, biais, 'o-', color='green', label='Biais', linewidth=2)
plt.plot(train_sizes, variance, 's-', color='orange', label='Variance', linewidth=2)
plt.title("Random Forest - Biais / Variance")
plt.xlabel("Nombre d'exemples d'entraînement")
plt.ylabel("Erreur")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Biais final : {biais[-1]:.4f} | Variance finale : {variance[-1]:.4f}")
```

```python
# LOGISTIC REGRESSION + BIAIS
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Entraîner Logistic Regression
lr_model = LogisticRegression(max_iter=1000, solver='lbfgs', random_state=42, n_jobs=-1)
lr_model.fit(X_train, y_train)

# Accuracy
train_acc = accuracy_score(y_train, lr_model.predict(X_train))
test_acc = accuracy_score(y_test, lr_model.predict(X_test))

print("LOGISTIC REGRESSION")
print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

# Courbe d'apprentissage - Biais/Variance
train_sizes, train_scores, val_scores = learning_curve(
    lr_model, X_train, y_train, cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy', n_jobs=-1
)

train_mean = np.mean(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Train', linewidth=2)
plt.plot(train_sizes, val_mean, 'o-', color='red', label='Validation', linewidth=2)
plt.title("Logistic Regression - Courbe d'Apprentissage")
plt.xlabel("Nombre d'exemples d'entraînement")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

biais = 1 - val_mean
variance = train_mean - val_mean

plt.figure(figsize=(8,5))
plt.plot(train_sizes, biais, 'o-', color='green', label='Biais', linewidth=2)
plt.plot(train_sizes, variance, 's-', color='orange', label='Variance', linewidth=2)
plt.title("Logistic Regression - Biais / Variance")
plt.xlabel("Nombre d'exemples d'entraînement")
plt.ylabel("Erreur")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Biais final : {biais[-1]:.4f} | Variance finale : {variance[-1]:.4f}")
```

```python
# SVM (SUPPORT VECTOR MACHINE) + BIAIS
from sklearn.svm import SVC
from sklearn.model_selection import learning_curve
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Entraîner SVM
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42, probability=True)
svm_model.fit(X_train, y_train)

# Accuracy
train_acc = accuracy_score(y_train, svm_model.predict(X_train))
test_acc = accuracy_score(y_test, svm_model.predict(X_test))

print("SUPPORT VECTOR MACHINE (SVM)")
print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")

train_sizes, train_scores, val_scores = learning_curve(
    svm_model, X_train, y_train, cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='accuracy', n_jobs=-1
)

train_mean = np.mean(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Train', linewidth=2)
plt.plot(train_sizes, val_mean, 'o-', color='red', label='Validation', linewidth=2)
plt.title("SVM - Courbe d'Apprentissage")
plt.xlabel("Nombre d'exemples d'entraînement")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

biais = 1 - val_mean
variance = train_mean - val_mean

plt.figure(figsize=(8,5))
plt.plot(train_sizes, biais, 'o-', color='green', label='Biais', linewidth=2)
plt.plot(train_sizes, variance, 's-', color='orange', label='Variance', linewidth=2)
plt.title("SVM - Biais / Variance")
plt.xlabel("Nombre d'exemples d'entraînement")
plt.ylabel("Erreur")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Biais final : {biais[-1]:.4f} | Variance finale : {variance[-1]:.4f}")
```

```python
# COMPARAISON DES MODÈLES - OBSERVATIONS ET CONCLUSION
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

print("COMPARAISON DES TROIS MODÈLES DE CLASSIFICATION")


rf_pred_train = rf_model.predict(X_train)
rf_pred_test  = rf_model.predict(X_test)

lr_pred_train = lr_model.predict(X_train)
lr_pred_test  = lr_model.predict(X_test)

svm_pred_train = svm_model.predict(X_train)
svm_pred_test  = svm_model.predict(X_test)


models_comparison = {
    'Random Forest': {
        'train_acc': accuracy_score(y_train, rf_pred_train),
        'test_acc': accuracy_score(y_test, rf_pred_test),
    },
    'Logistic Regression': {
        'train_acc': accuracy_score(y_train, lr_pred_train),
        'test_acc': accuracy_score(y_test, lr_pred_test),
    },
    'SVM': {
        'train_acc': accuracy_score(y_train, svm_pred_train),
        'test_acc': accuracy_score(y_test, svm_pred_test),
    }
}

comparison_df = pd.DataFrame({
    'Modèle': list(models_comparison.keys()),
    'Train Accuracy': [models_comparison[m]['train_acc'] for m in models_comparison],
    'Test Accuracy':  [models_comparison[m]['test_acc']  for m in models_comparison],
})

comparison_df['Différence (Train-Test)'] = (
    comparison_df['Train Accuracy'] - comparison_df['Test Accuracy']
)
comparison_df['Overfitting'] = comparison_df['Différence (Train-Test)'].apply(
    lambda x: 'Oui' if x > 0.1 else 'Non'
)

print("\nTABLE DE COMPARAISON DES MÉTRIQUES\n")
print(comparison_df.to_string(index=False))

plt.figure(figsize=(8,5))

x = np.arange(len(comparison_df))
width = 0.35

plt.bar(x - width/2, comparison_df['Train Accuracy'],
        width, label='Train', alpha=0.8)
plt.bar(x + width/2, comparison_df['Test Accuracy'],
        width, label='Test', alpha=0.8)

plt.xticks(x, comparison_df['Modèle'], rotation=15)
plt.ylabel("Accuracy")
plt.title("Comparaison Train vs Test Accuracy")
plt.ylim(0, 1)
plt.legend()
plt.grid(axis='y', alpha=0.3)

for i, (train, test) in enumerate(zip(
        comparison_df['Train Accuracy'], comparison_df['Test Accuracy'])):
    plt.text(i - width/2, train + 0.02, f"{train:.3f}", ha='center')
    plt.text(i + width/2, test + 0.02, f"{test:.3f}", ha='center')

plt.show()

plt.figure(figsize=(8,5))

colors = ['red' if x > 0.1 else 'green'
          for x in comparison_df['Différence (Train-Test)']]

plt.bar(comparison_df['Modèle'],
        comparison_df['Différence (Train-Test)'],
        color=colors, alpha=0.8)

plt.axhline(y=0.1, linestyle='--', linewidth=2,
            label='Seuil overfitting (0.1)')

plt.ylabel("Différence (Train - Test)")
plt.title("Indicateur d’Overfitting")
plt.legend()
plt.grid(axis='y', alpha=0.3)

for i, diff in enumerate(comparison_df['Différence (Train-Test)']):
    plt.text(i, diff + 0.01, f"{diff:.3f}", ha='center')

plt.show()

print("\nOBSERVATIONS PAR MODÈLE")

for i, row in comparison_df.iterrows():
    print(f"\n{row['Modèle'].upper()}")
    print(f"   Train Accuracy : {row['Train Accuracy']:.4f}")
    print(f"   Test Accuracy  : {row['Test Accuracy']:.4f}")
    print(f"   Différence     : {row['Différence (Train-Test)']:.4f}")

    if row['Différence (Train-Test)'] > 0.1:
        print("   Observation : Overfitting détecté")
    elif row['Train Accuracy'] < 0.75:
        print("   Observation : Underfitting (modèle trop simple)")
    else:
        print("   Observation : Bon équilibre biais/variance")

best_idx = comparison_df['Test Accuracy'].idxmax()
best_model = comparison_df.loc[best_idx]

print("\nCONCLUSION ET RECOMMANDATION FINALE")
print(f"\nMEILLEUR MODÈLE : {best_model['Modèle']}")
print(f"Accuracy Test  : {best_model['Test Accuracy']:.4f}")

if best_model['Modèle'] == 'Random Forest':
    print(" Excellentes performances en test")
    print("Léger overfitting possible")
    print("Recommandé pour la production avec régularisation fine")
elif best_model['Modèle'] == 'SVM':
    print("Très bon compromis biais/variance")
    print("Bonne généralisation")
    print("Modèle recommandé pour la production")
else:
    print("Modèle linéaire trop simple")
    print("Non recommandé pour la production")
```
