```python
import pandas as pd
import os

DATA_DIR = "condition+monitoring+of+hydraulic+systems"

# Fichiers Ã  exclure pour le chargement 
EXCLUDED_FILES = {
    "profile.txt",
    "description.txt",
    "documentation.txt"
}

# Fonction de chargement des fichiers de capteurs
def load_and_rename_sensor(file_path):
    """
    Charge un fichier capteur (PS1, FS1, TS1, etc.)
    et renomme les colonnes : NomCapteur_1, NomCapteur_2, ...
    """
    sensor_name = os.path.splitext(os.path.basename(file_path))[0]

    df = pd.read_csv(
        file_path,
        sep=r"\s+",
        header=None,
        encoding="latin1"
    )

    df.columns = [f"{sensor_name}_{i+1}" for i in range(df.shape[1])]
    return df

# Chargement des fichiers capteurs uniquement
sensor_dfs = []

for file in sorted(os.listdir(DATA_DIR)):
    file_lower = file.lower()

    if (
        file_lower.endswith(".txt")
        and file_lower not in EXCLUDED_FILES
    ):
        file_path = os.path.join(DATA_DIR, file)
        df_sensor = load_and_rename_sensor(file_path)
        sensor_dfs.append(df_sensor)
        print(f"âœ” ChargÃ© : {file}")

# Chargement du fichier profile contenant la variable cible
profile_df = pd.read_csv(
    os.path.join(DATA_DIR, "profile.txt"),
    sep=r"\s+",
    header=None,
    encoding="latin1"
)

profile_df.columns = [
    "cooler_condition",
    "valve_condition",
    "internal_pump_leakage",
    "hydraulic_accumulator",
    "stable_flag"
]

# Fusion finale
final_df = pd.concat(sensor_dfs + [profile_df], axis=1)

print("\n Fusion terminÃ©e")
print("Shape finale :", final_df.shape)
print("\nDerniÃ¨res colonnes (targets) :")
print(final_df.columns[-5:])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("ðŸ”¹ Shape du dataset :", final_df.shape)
print("\nðŸ”¹ Types de variables :")
print(final_df.dtypes.value_counts())


# VÃ©rification des valeurs manquantes
missing_count = final_df.isnull().sum()
missing_percent = (missing_count / len(final_df)) * 100

missing_df = pd.DataFrame({
    "missing_count": missing_count,
    "missing_percent": missing_percent
})

# Conserver uniquement les colonnes avec NaN
missing_df = missing_df[missing_df["missing_count"] > 0]

print("\nðŸ”¹ Colonnes avec valeurs manquantes :")
print(missing_df)

if missing_df.empty:
    print("Aucune valeur manquante dÃ©tectÃ©e")


# DÃ©tection des outliers (mÃ©thode IQR â€“ globale)
numeric_df = final_df.select_dtypes(include=[np.number])

outliers_summary = []

for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1

    if IQR == 0:
        continue

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers_count = ((numeric_df[col] < lower) | (numeric_df[col] > upper)).sum()

    if outliers_count > 0:
        outliers_summary.append({
            "variable": col,
            "outliers_count": outliers_count,
            "outliers_percent": (outliers_count / len(numeric_df)) * 100
        })

outliers_df = pd.DataFrame(outliers_summary)

print("\nðŸ”¹ Nombre de variables contenant des outliers :", outliers_df.shape[0])
print(outliers_df.head())

# Distribution du pourcentage d'outliers par variables
plt.figure(figsize=(8,4))
plt.hist(outliers_df["outliers_percent"], bins=30)
plt.title("Distribution du % d'outliers par variable")
plt.xlabel("% d'outliers")
plt.ylabel("Nombre de variables")
plt.show()

sample_cols = outliers_df.sort_values(
    "outliers_percent", ascending=False
)["variable"].head(5)

plt.figure(figsize=(12,6))
numeric_df[sample_cols].boxplot()
plt.xticks(rotation=45)
plt.title("Boxplots â€“ variables avec le plus d'outliers")
plt.show()
```
