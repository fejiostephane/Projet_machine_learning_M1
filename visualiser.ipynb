```python
# Méthode 1 : Découpage en fenêtres + statistiques pour PS1

# Charger PS1
ps1_df = pd.read_csv("PS1.txt", sep=r"\s+", header=None)

# Prendre la première ligne (6000 points)
ps1_signal = ps1_df.iloc[0].values

print(f"Signal original : {len(ps1_signal)} points")

# Paramètres de fenêtrage
window_size = 100  # 100 points par fenêtre
n_windows = len(ps1_signal) // window_size  # 60 fenêtres

# Découper en fenêtres
windows = ps1_signal[:n_windows * window_size].reshape(n_windows, window_size)

print(f"Nombre de fenêtres : {n_windows}")
print(f"Taille de chaque fenêtre : {window_size} points")

# Calculer les statistiques pour chaque fenêtre
stats_mean = np.mean(windows, axis=1)  # 60 moyennes
stats_max = np.max(windows, axis=1)    # 60 max
stats_min = np.min(windows, axis=1)    # 60 min
stats_std = np.std(windows, axis=1)    # 60 écart-types
stats_median = np.median(windows, axis=1)  # 60 médianes

# Créer un DataFrame avec toutes les statistiques
stats_df = pd.DataFrame({
    'moyenne': stats_mean,
    'max': stats_max,
    'min': stats_min,
    'ecart_type': stats_std,
    'mediane': stats_median
})

print(f"\n✅ Réduction : {len(ps1_signal)} → {len(stats_df) * 5} valeurs")
print(f"   ({len(stats_df)} fenêtres × 5 statistiques)")

stats_df
```

```python
# Visualiser la réduction de données

fig, axes = plt.subplots(3, 1, figsize=(14, 12))

# Signal original (6000 points)
axes[0].plot(ps1_signal, linewidth=0.5, alpha=0.7)
axes[0].set_title(f"Signal original PS1 - {len(ps1_signal)} points", fontsize=14, fontweight='bold')
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Valeur")
axes[0].grid(True, alpha=0.3)

# Signal réduit (moyenne par fenêtre - 60 points)
x_reduced = np.arange(len(stats_mean)) * window_size + window_size/2  # Centre de chaque fenêtre
axes[1].plot(x_reduced, stats_mean, marker='o', linewidth=2, markersize=4, label='Moyenne')
axes[1].fill_between(x_reduced, stats_min, stats_max, alpha=0.3, label='Min-Max')
axes[1].set_title(f"Signal réduit - {len(stats_mean)} fenêtres (moyenne + enveloppe min-max)", fontsize=14, fontweight='bold')
axes[1].set_xlabel("Index (centre de fenêtre)")
axes[1].set_ylabel("Valeur")
axes[1].grid(True, alpha=0.3)
axes[1].legend()

# Troisième graphique : médiane + enveloppe min-max
axes[2].fill_between(x_reduced, stats_min, stats_max, alpha=0.25, label='Min-Max')
axes[2].plot(x_reduced, stats_median, marker='o', linewidth=2, markersize=4, color='C2', label='Médiane')
axes[2].set_title("Médiane par fenêtre avec enveloppe Min-Max", fontsize=14, fontweight='bold')
axes[2].set_xlabel("Index (centre de fenêtre)")
axes[2].set_ylabel("Valeur")
axes[2].grid(True, alpha=0.3)
axes[2].legend()

plt.tight_layout()
plt.show()

print(f"\n📊 Résumé :")
print(f"   • Signal original : {len(ps1_signal)} points")
print(f"   • Signal réduit : {len(stats_mean)} fenêtres")
print(f"   • Taux de compression : {len(ps1_signal) / len(stats_mean):.1f}x")
print(f"   • Total features (5 stats × {len(stats_mean)} fenêtres) : {len(stats_mean) * 5} valeurs")
```

D’accord. Là tu as zoomé sur **les 0 à 300 premiers points** de PS1.
Comme PS1 est à **100 Hz**, ça correspond à environ **0 à 3 secondes** du cycle (300 points ÷ 100 = 3 s). 

Je t’explique **exactement** ce qu’on voit, **phase par phase**, et pourquoi c’est intéressant (surtout pour la valve).

---

## 1) Ce que représente ce zoom

* **Axe X (Index)** : le temps *dans le cycle* (ici les toutes premières secondes)
* **Axe Y (Valeur)** : la pression mesurée par PS1 (en bar)

Ce bout du signal est souvent une phase où :

* la machine “démarre” un état de charge,
* des éléments (dont la valve) changent de position,
* la pression se stabilise sur un niveau cible.

---

## 2) Décomposition du signal : 4 phases visibles

### Phase A — tout début (index ~0 à ~10)

On voit une pression autour de **150–155** avec une petite irrégularité au démarrage.

👉 Interprétation simple :

* démarrage / mise en régime
* petit “à-coup” normal de début de cycle (bruit, commutation, etc.)

📌 Ce que ça “dit” au ML :

* **comment le cycle commence** (important si la valve a du retard ou répond mal)

---

### Phase B — montée progressive (index ~10 à ~70)

La pression **monte presque en ligne droite** jusque vers **190–191**.

👉 Ça ressemble à une **rampe** :

* le système “pousse” de plus en plus fort
* la pression augmente de manière régulière

📌 Pourquoi c’est important pour la valve :

* une valve **lente** ou avec **retard** peut changer la **vitesse de montée**
* donc la **pente** de cette rampe est une info très utile

✅ Feature typique ici :

* **pente (slope)** = (pression fin – pression début) / durée

---

### Phase C — plateau stable haut (index ~70 à ~170)

La pression devient **presque constante** autour de **191**.

👉 Interprétation simple :

* le système a atteint un “mode stable”
* tout est en équilibre pendant un moment

📌 Ce que ça dit :

* la **stabilité** (faible bruit) → système sain / bien contrôlé
* si la valve fuit / répond mal → plateau moins stable, fluctuations plus fortes

✅ Features typiques :

* **moyenne** sur la fenêtre
* **écart-type** (stabilité)
* **min–max** (amplitude)

---

### Phase D — chute très brutale puis remontée (index ~170 à ~230+)

C’est la partie la plus “spectaculaire” :

* une **chute quasi verticale** de ~191 vers ~162
* puis une **remontée progressive** vers ~191

👉 Interprétation simple :

* il se passe un **événement** (commutation / ouverture-fermeture / changement d’état)
* la pression “tombe” puis le système “récupère” et remonte

📌 Pourquoi c’est *très* important pour ton sujet “valve” :

* une valve, c’est justement un composant qui **commute** (change de position)
* si elle est lente, tu peux avoir :

  * chute plus longue
  * remontée plus lente
  * retard avant la remontée
  * oscillations après la commutation

✅ Features typiques pour cette zone :

* **profondeur de chute** = plateau haut − minimum après chute
* **temps de récupération** = combien de temps pour revenir proche du plateau
* **pente de remontée**
* **aire sous la courbe** pendant l’événement (optionnel)

---

## 3) Comment relier ça à ton “découpage en fenêtres”

Si tu découpes en fenêtres de **100 points (1 seconde)** :

* Fenêtre 1 (0–100) : montée + début plateau
* Fenêtre 2 (100–200) : plateau + chute brutale
* Fenêtre 3 (200–300) : remontée + plateau

👉 C’est exactement pour ça que **moyenne/min/max** marchent bien :

* la fenêtre qui contient la chute aura un **min très bas** et un **max très haut**
* ça “signale” un événement de commutation

📌 Mais attention :

* si la chute est très courte et tombe “entre” deux fenêtres, tu peux l’atténuer
  ➡️ D’où l’intérêt d’ajouter :
* **écart-type** (std)
* **pente**
  ou d’utiliser des fenêtres un peu plus petites (ex 50 points = 0,5s)

---

## 4) Ce que tu dois retenir

Dans ce zoom, tu as 3 infos très utiles pour classifier la valve :

1. **La vitesse de montée** (rampe)
2. **La stabilité du plateau**
3. **La signature de commutation** (chute + récupération)

👉 Si la valve est en retard (classes 90, 80, 73), ces 3 éléments changent souvent.

```python
# Option A : Concaténer les features de PS1..PS6 pour chaque cycle
import pandas as pd
import numpy as np

# Paramètres
window_size = 100  # 100 points = 1 seconde
stats_order = ("mean", "min", "max", "std")  # 4 statistiques

sensors = ["PS1", "PS2", "PS3", "PS4", "PS5", "PS6"]
dfs = {s: pd.read_csv(f"{s}.txt", sep=r"\s+", header=None) for s in sensors}

# Vérifications de cohérence
n_cycles = min(df.shape[0] for df in dfs.values())
signal_length = min(dfs[s].shape[1] for s in sensors)
n_windows = signal_length // window_size  # attendu 6000/100 = 60

print(f"Capteurs: {sensors}")
print(f"Cycles disponibles: {n_cycles}")
print(f"Longueur du signal: {signal_length} points")
print(f"Nombre de fenêtres: {n_windows} (taille {window_size})")

def compute_window_stats(signal: np.ndarray, window_size: int):
    # Tronquer pour avoir un multiple entier de window_size
    usable = signal[: (len(signal) // window_size) * window_size]
    windows = usable.reshape(-1, window_size)
    stats = {
        "mean": windows.mean(axis=1),
        "min": windows.min(axis=1),
        "max": windows.max(axis=1),
        "std": windows.std(axis=1),
    }
    return stats  # chaque entrée: vecteur de longueur n_windows

# Construire les noms de colonnes dans le même ordre que la concaténation
columns = []
for s in sensors:
    for stat in stats_order:
        columns.extend([f"{s}_{stat}_w{w}" for w in range(n_windows)])

# Extraire les features pour tous les cycles
features_rows = []
for cycle_idx in range(n_cycles):
    row_features = []
    for s in sensors:
        signal = dfs[s].iloc[cycle_idx].values
        stats = compute_window_stats(signal, window_size)
        # Concaténer dans l'ordre défini par stats_order
        for stat in stats_order:
            row_features.extend(stats[stat].tolist())
    features_rows.append(row_features)

features_df = pd.DataFrame(features_rows, columns=columns)
features_df.insert(0, "cycle", np.arange(1, n_cycles + 1))

print(f"\n✅ X_final_par_cycle: shape = {features_df.shape} (cycles × features)")
print(f"   • Features par capteur: {n_windows} fenêtres × {len(stats_order)} stats = {n_windows * len(stats_order)}")
print(f"   • Total features (6 capteurs): {(n_windows * len(stats_order)) * len(sensors)}")

# Exemple demandé: prendre la ligne 100 (1-based) pour PS1..PS6 concaténées
cycle_number = 100
if cycle_number <= n_cycles:
    example_row = features_df[features_df["cycle"] == cycle_number].iloc[0]
    print(f"\nExemple: Features concaténées pour le cycle {cycle_number}")
    display(example_row.to_frame().T)
else:
    print(f"\nCycle {cycle_number} > n_cycles ({n_cycles}) — ajustez cycle_number.")

# Aperçu global
features_df.head(3)
```

```python
# Visualisation pour tous les PS : original + moyenne avec enveloppe min-max
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sensors = ["PS1", "PS2", "PS3", "PS4", "PS5", "PS6"]
dfs = {s: pd.read_csv(f"{s}.txt", sep=r"\s+", header=None) for s in sensors}

# Choisir le cycle à visualiser (1-based)
cycle_number = 1
cycle_idx = cycle_number - 1

# Paramètres de fenêtrage
window_size = 100

stats_per_sensor = {}
original_per_sensor = {}

for s in sensors:
    signal = dfs[s].iloc[cycle_idx].values
    original_per_sensor[s] = signal
    n_windows = len(signal) // window_size
    usable = signal[: n_windows * window_size]
    windows = usable.reshape(n_windows, window_size)
    stats_mean = windows.mean(axis=1)
    stats_min = windows.min(axis=1)
    stats_max = windows.max(axis=1)
    x_reduced = np.arange(n_windows) * window_size + window_size / 2  # centres de fenêtres
    stats_per_sensor[s] = {
        "mean": stats_mean,
        "min": stats_min,
        "max": stats_max,
        "x": x_reduced,
        "n_windows": n_windows,
    }

# Figure 1 : Signaux originaux (PS1..PS6)
fig, axes = plt.subplots(3, 2, figsize=(16, 10), sharex=False)
for i, s in enumerate(sensors):
    ax = axes[i // 2, i % 2]
    sig = original_per_sensor[s]
    ax.plot(sig, linewidth=0.6)
    ax.set_title(f"{s} - Signal original ({len(sig)} points)")
    ax.set_xlabel("Index")
    ax.set_ylabel("Valeur")
    ax.grid(True, alpha=0.3)
fig.suptitle(f"Signals originaux - Cycle {cycle_number}", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()

# Figure 2 : Statistiques par fenêtres (moyenne + enveloppe min-max)
fig, axes = plt.subplots(3, 2, figsize=(16, 10), sharex=False)
for i, s in enumerate(sensors):
    ax = axes[i // 2, i % 2]
    d = stats_per_sensor[s]
    ax.fill_between(d["x"], d["min"], d["max"], alpha=0.25, label="Min-Max")
    ax.plot(d["x"], d["mean"], marker="o", linewidth=2, markersize=3, label="Moyenne")
    ax.set_title(f"{s} - Statistiques par fenêtre (n={d['n_windows']})")
    ax.set_xlabel("Index (centre de fenêtre)")
    ax.set_ylabel("Valeur")
    ax.grid(True, alpha=0.3)
    ax.legend()
fig.suptitle(f"Réduction par fenêtres - Cycle {cycle_number}", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()

print("\n📊 Infos :")
print(f"   • Cycle visualisé : {cycle_number}")
print(f"   • Fenêtrage : {window_size} points (≈ 1s si 100 Hz)")
```
